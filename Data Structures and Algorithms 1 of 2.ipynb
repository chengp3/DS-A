```python
def findint(arr, num):
    start = 0
    mid = round(len(arr)/2)
    end = len(arr)-1
    
    index = -1
    # case where arr = [0,1,2,3,4,5,7] and num = 6
    # start = 6
    # mid = 7
    # end = 7
    
    if num < arr[start] or num > arr[end]:
        return "not in array"
    
    while (index == -1): 
        print(arr[start],arr[mid],arr[end])
        
        if arr[mid] == num:
            index = mid
        elif ((start == mid-1) and (arr[start]==num)):
            index = start
        elif ((mid==end) and (arr[mid] != num)):
            return "not in array"
        elif (num >= arr[start]) and (num <= arr[mid]):
            end = mid
            mid = round((mid-start)/2) + start
        elif (num >= arr[mid]) and (num <= arr[end]):
            start = mid
            mid = math.ceil((end-mid)/2) + mid
            
    return index
```

# Bubble sort
look at two values, if they're out of order swap them

move the frame one to the right. repeat

reach the end, go back to the beginning and repeat until you don't do any switches

```python
arr = random.sample(range(10),6)
print(arr)
```

```python
def passthrough(arr):
    i = 0
    j = 1
    clean = 1

    while j < len(arr):
        if arr[i] > arr[j]:
            temp = arr[i]
            arr[i] = arr[j]
            arr[j] = temp
            clean = 0
        i += 1
        j += 1
    return arr, clean
        
def bubble_sort(arr):
    
    clean = 0
        
    if len(arr) <= 2: return "array not big enough"
    
    while clean == 0:
        arr, clean = passthrough(arr)
        print(arr, clean)
    return arr
        
```

# Selection Sort

```python
import random
sample = random.sample(range(10),6)

# Start with first element, go right until you find lowest element. Swap them.
# Repeat with every element.

def selection_sort(arr):
    print(f'original: {arr}')
    i = 0
    while i < (len(arr)-1): # i = passthrough number, for array of 5 will have i = 0,1,2,3
        print(f'i: {i}')
        j = i+1
        low_index = i
        while j < (len(arr)): # j = next after i, j = 1,2,3,4
            print(f'j: {j}')
            if arr[j] < arr[low_index]:
                low_index = j
            j += 1
        print(f'low_index: {low_index} = arr[low_index]')
        temp = arr[i]
        arr[i] = arr[low_index]
        arr[low_index] = temp
        print(f'passthrough {i+1}: {arr}')
               
        i += 1
    return arr

arr = selection_sort(sample)

print(f'final result: {arr}')
```

# Hash maps

Reduced lookup from N-squared or log-N to O(1). 

Use it as a one-directional lookup in any situation. For example, for any list you reference simply to check the existence of that item... { 1 : true, 2 : true, 3 : true...}

python hash(whatever) here to generate hash. it's pretty regular which means integers evaluate to the same integers. however non-integers also map to integers, which means collisions are inevitable (while rare). python deals with this by having "open addressing" which essentially means it continues down the hash in a predictable way until it finds an open spot. For lookup, it does the same thing, finding the first index before going down to the open spot in the same way, before retrieving the value.

What is a bottom-up algorithm?

Commonly a loop. Starting with the small, simple component or base case and adding onto it. 

Programming wise this can be a simple loop OR recursion but recursion doesn't add anything extra and is not intuitive.

What is a top-down algorithm?

Starting with the most complicated, and reduce complexity by looking at the nearest sub-problem. Recursion is the only way to accomplish this.
