# Linked lists



```python
class LinkedList:
    def __init__(self, nodes=None):
        if nodes != None:
            node = Node(data = nodes[0])
            self.head = node
            i = 1
            while i < len(nodes):
                node.next = Node(data = nodes[i])
                node = node.next
                i += 1

    def __iter__(self):
        node = self.head
        while node is not None:
            yield node
            node = node.next            
    
    def indexof(self, search_value):
        node = self.head
        i = 0
        
        if node.data == search_value:
            return i
        
        node = node.next
        
        while node != None:
            if node.data == search_value:
                return i
            node = node.next
            i += 1
        return "Not found"
    
    def __repr__(self):
        node = self.head
        nodes = []
        while node is not None:
            nodes.append(str(node.data))
            node = node.next
        nodes.append("END")
        return " -> ".join(nodes)
    
    #method to read
    
    def read(self, index):
        i = 0
        node = self.head
        while i != index:
            node = node.next
            i += 1
        return node.data
    
    #method to insert
    
    def insert(self, index, value):
        i = 0
        node = self.head
        
        
        if self.head == None:
            self.head = Node(value)
            return
        
        ref = node.next
        
        while i != (index-1):
            node = node.next
            ref = node.next
            i += 1
        
        node.next = Node(value)
        node = node.next
        node.next = ref
        
    
    #method to delete
    
    def delete(self, index):
        i = 0
        beforenode = None
        node = self.head
        ref = None
        
        if (index == 0) and (node.next == None):
            self.head = None
            return
        
        if (index == 0) and (node.next != None):
            self.head = node.next
            return
        
        while i != index:
            beforenode = node
            node = node.next
            
            i += 1
        
        ref = node.next # get the deleting node's reference to set beforenode.next
        beforenode.next = ref # set it
        
        del node
        
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

    def __repr__(self):
        return f'{self.data}'
    

llist = LinkedList([2,4,6,8])
print(llist)

llist.insert(2,10)
print(llist.read(2))

print(llist)

llist.delete(0)
llist.delete(0)
llist.delete(0)
llist.delete(0)
llist.delete(0)
llist.insert(0,3)
print(llist)

print(llist.indexof(3))
```

# Binary Tree

```python
class tnode():
    def __init__(self, data, l = None, r = None):
        self.data = data
        self.left = l
        self.right = r

    def __repr__(self):
        return f'{self.data}'
    
n1 = tnode(5)
n2 = tnode(8)
n3 = tnode(4)
n4 = tnode(1)
n5 = tnode(2)
n6 = tnode(9)
n7 = tnode(10)

#     5  
#    4 8
#   1   9
#    2   10

n1.right = n2
n1.left = n3
n3.left = n4
n4.right = n5
n2.right = n6
n6.right = n7
```

```python
# print everything

def inorder(node):
    if node.left:
        printall(node.left)
    print(node.data)
    if node.right:
        printall(node.right)

def preorder(node):
    print(node.data)    
    if node.left:
        preorder(node.left)
    if node.right:
        preorder(node.right)

def postorder(node):
    if node == None:
        return
    postorder(node.left)
    postorder(node.right)
    print(node.data)
        
root = n1
postorder(n1)
```

```python
# search the tree
root = n1

def search(value, node):
    print(node.data)
    if value == node.data:
        print("found")
        return
    elif (node.left == None) and (node.right == None):
        print("Not found")
        return
    if value < node.data:
        search(value,node.left)
    if value >= node.data:
        search(value,node.right)
    
search(2, n1)
        
```

```python
# insert node
# descend tree until it ends, insert it 
new = tnode(6)
root = n1

def insert(new, node):
    if (new.data < node.data) and (node.left == None):
        node.left = new
        return
    elif (new.data >= node.data) and (node.right == None):
        node.right = new
        return
    if (new.data < node.data) and (node.left != None):
        insert(new,node.left)
    if (new.data >= node.data) and (node.right != None):
        insert(new,node.right)
        
insert(new, root)
print(n2.left)
```

```python
# delete node
# if no children, simply delete
# if one child, replace node with child
# if two children, replace node with "successor" node. 
#    successor = go right one then left all the way.
#        if successor has a right child, replace successor with the right child (becomes left child of successor's parent)
```

```python
# invert the tree

def invert(node):
    if node:
        temp = node.left 
        node.left = node.right
        node.right = temp

        invert(node.left)
        invert(node.right)
        
# O(N) reversion, O(N) iterative as well
```

### Notes
Traversal is always in the same order but the order of official registration is different. 

Deletion and Insertion are both O(logN), search is O(logN). Overall very fast especially for applications where you have to change data a lot.
